## Imports

```python
import time
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV, cross_validate
from sklearn.preprocessing import MinMaxScaler
from joblib import dump

from aqosd_experiments.config import CLASSIFIERS, PARAM_GRIDS, RAW_DATASET_PATH, HOST_LIST, CV, MODELS_PATH, FIG_PATH
from aqosd_experiments.data import import_and_prepare_data, scale_metrics, over_sampling
from aqosd_experiments.plot import plot_number_of_instance, plot_osdm, plt_long_stats, plt_corr_metrics, \
    plot_multicollinear_metrics, plt_corr_bottlenecks
from aqosd_experiments.scorers import process_score, SCORING
from osms import OverheadSensitiveMetricSelection
```

## Load Config

```python
raw_dataset_path, host_list, models_path, fig_path= RAW_DATASET_PATH,  HOST_LIST, MODELS_PATH, FIG_PATH
classifiers,param_grids = CLASSIFIERS, PARAM_GRIDS
scoring, cv = SCORING, CV
over_sampling_factor=0.1
save=True
```

## Load and prepare data

```python
fig = plt_long_stats(raw_dataset_path, host_list)
if save:
    fig.savefig(fig_path + "Sample.pdf", bbox_inches='tight')
```

```python
metrics, bottlenecks = import_and_prepare_data(raw_dataset_path, host_list)
print('Shape of metrics : ',metrics.shape,'\t','Shape of bottlenecks : ',bottlenecks.shape)
print('Label cardinality = %.2f \t Label density = %.2f' % (bottlenecks.sum(axis=1).mean(),
                                                           bottlenecks.mean(axis=1).mean()))
```

```python
metrics=scale_metrics(metrics, MinMaxScaler()) 
print('Shape of metrics : ',metrics.shape,'\t','Shape of bottlenecks : ',bottlenecks.shape)
```

```python
metrics, bottlenecks = over_sampling(metrics, bottlenecks, int(metrics.shape[0]*over_sampling_factor))
print('Shape of metrics : ',metrics.shape,'\t','Shape of bottlenecks : ',bottlenecks.shape)
print('Label cardinality = %.2f \t Label density = %.2f' % (bottlenecks.sum(axis=1).mean(),bottlenecks.mean(axis=1).mean()))
```

```python
metric_names, bottleneck_names = list(metrics.columns), list(bottlenecks.columns)
print(metric_names)
print(100*'-')
print(bottleneck_names)
```

```python
from sklearn.ensemble import RandomForestClassifier
from skmultilearn.model_selection import iterative_train_test_split
import matplotlib.pyplot as plt
from sklearn.inspection import permutation_importance
import numpy as np

X_train, y_train, X_test, y_test = iterative_train_test_split(metrics.values, bottlenecks.values, test_size = 0.25)
rf = RandomForestClassifier(random_state=42,n_estimators=100,n_jobs=3)
rf.fit(X_train, y_train)
```

```python
print("RF train accuracy: %0.3f" % rf.score(X_train, y_train))
print("RF test accuracy: %0.3f" % rf.score(X_test, y_test))
```

```python
result = permutation_importance(rf, X_train, y_train, n_repeats=20, random_state=42)
```

```python
perm_sorted_idx = result.importances_mean.argsort()
tree_importance_sorted_idx = np.argsort(rf.feature_importances_)
tree_indices = np.arange(0, len(rf.feature_importances_)) + 0.5
feature_names= np.array(metric_names)
```

```python
fig, (ax2, ax1) = plt.subplots(1, 2, figsize=(6, 12), sharey=False, gridspec_kw={'width_ratios': [5, 1]})
ax1.barh(tree_indices, rf.feature_importances_[tree_importance_sorted_idx], height=0.5)
ax1.set_yticklabels(feature_names[tree_importance_sorted_idx])
ax1.set_yticks(tree_indices)
ax1.set_ylim((0, len(rf.feature_importances_)))
ax2.boxplot(result.importances[perm_sorted_idx].T, vert=False, labels=feature_names[perm_sorted_idx])
for ax in (ax1, ax2):
    [item.set_fontsize(6) for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] + ax.get_xticklabels() + ax.get_yticklabels())]
fig.tight_layout()
plt.show()
```
